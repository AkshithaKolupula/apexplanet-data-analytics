# SQL Fundamentals

view first 10 rows

```python
query = """
SELECT *
FROM cleaned_superstore
LIMIT 10;
"""

pd.read_sql(query, engine)
```

Sales greater than 1000

```python
query = """
SELECT *
FROM cleaned_superstore
WHERE Sales > 1000;
"""

pd.read_sql(query, engine)
```

Top 10 highest sales

```python
query = """
SELECT *
FROM cleaned_superstore
ORDER BY Sales DESC
LIMIT 10;
"""

pd.read_sql(query, engine)
```

Total Sales by Category

```python
query = """
SELECT Category,
SUM(Sales) AS Total_Sales
FROM cleaned_superstore
GROUP BY Category;
"""

pd.read_sql(query, engine)
```

Categories having sales above ₹500000

```python
query = """
SELECT Category,
SUM(Sales) AS Total_Sales
FROM cleaned_superstore
GROUP BY Category
HAVING SUM(Sales) > 500000;
"""

pd.read_sql(query, engine)
```

Top 5 Products by Sales (Business Question)

```python
query = """
SELECT `Product Name`,
SUM(Sales) AS Total_Sales
FROM cleaned_superstore
GROUP BY `Product Name`
ORDER BY Total_Sales DESC
LIMIT 5;
"""

pd.read_sql(query, engine)
```

Monthly Sales Trend

```python
query = """
SELECT
MONTH(`Order Date`) AS Month,
SUM(Sales) AS Total_Sales
FROM cleaned_superstore
GROUP BY MONTH(`Order Date`)
ORDER BY Month;
"""

pd.read_sql(query, engine)
```

Customer Spending

```python
query = """
SELECT
`Customer Name`,
SUM(Sales) AS Total_Spent
FROM cleaned_superstore
GROUP BY `Customer Name`
ORDER BY Total_Spent DESC
LIMIT 10;
"""

pd.read_sql(query, engine)
```

SUBQUERY

```python
query = """
SELECT `Product Name`, Sales
FROM cleaned_superstore
WHERE Sales > (
    SELECT AVG(Sales)
    FROM cleaned_superstore
)
ORDER BY Sales DESC;
"""

pd.read_sql(query, engine)
```

CTE (WITH Clause)

```python
query = """
WITH CategorySales AS (
    SELECT Category,
           SUM(Sales) AS Total_Sales
    FROM cleaned_superstore
    GROUP BY Category
)

SELECT *
FROM CategorySales
ORDER BY Total_Sales DESC;
"""

pd.read_sql(query, engine)
```

ROW_NUMBER()

```python
query = """
SELECT
    `Product Name`,
    Category,
    Sales,
    ROW_NUMBER() OVER(
        ORDER BY Sales DESC
    ) AS Row_Num
FROM cleaned_superstore;
"""

pd.read_sql(query, engine)
```

RANK()

```python
query = """
SELECT
    `Product Name`,
    Sales,
    RANK() OVER(
        ORDER BY Sales DESC
    ) AS Sales_Rank
FROM cleaned_superstore;
"""

pd.read_sql(query, engine)
```

LAG()

```python
query = """
SELECT
    `Order Date`,
    Sales,
    LAG(Sales) OVER(
        ORDER BY `Order Date`
    ) AS Previous_Sales
FROM cleaned_superstore;
"""

pd.read_sql(query, engine)
```

LEAD()

```python
query = """
SELECT
    `Order Date`,
    Sales,
    LEAD(Sales) OVER(
        ORDER BY `Order Date`
    ) AS Next_Sales
FROM cleaned_superstore;
"""

pd.read_sql(query, engine)
```

Create a View

```python
query = """
CREATE VIEW category_sales AS

SELECT
    Category,
    SUM(Sales) AS Total_Sales
FROM cleaned_superstore
GROUP BY Category;
"""

with engine.begin() as conn:
    conn.execute(text(query))

print("✅ View Created Successfully!")
```

View the View

```python
query = """
SELECT *
FROM category_sales;
"""

pd.read_sql(query, engine)
```

1. JOIN Example

```python
query = """
CREATE TABLE customer_info (
    `Customer ID` VARCHAR(20) PRIMARY KEY,
    Email VARCHAR(100)
);
"""

with engine.begin() as conn:
    conn.execute(text(query))

print("Customer table created.")
```

INNER JOIN

```python
query = """
SELECT
cs.`Customer Name`,
ci.Email,
cs.Sales

FROM cleaned_superstore cs

INNER JOIN customer_info ci
ON cs.`Customer ID` = ci.`Customer ID`;
"""

pd.read_sql(query, engine)
```
